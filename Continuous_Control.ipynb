### 2. Examine the State and Action Spaces

In this environment, a double-jointed arm can move to target locations. A reward of `+0.1` is provided for each step that the agent's hand is in the goal location. Thus, the goal of your agent is to maintain its position at the target location for as many time steps as possible.

The observation space consists of `33` variables corresponding to position, rotation, velocity, and angular velocities of the arm.  Each action is a vector with four numbers, corresponding to torque applicable to two joints.  Every entry in the action vector must be a number between `-1` and `1`.

Run the code cell below to print some information about the environment.

````python
def run_ddpg_for_one_episode(agent, env, brain_name, is_training, learn_every=1):
    """Run the DDPG agent for one episode"""
    env_info = env.reset(train_mode=is_training)[brain_name]   # reset the environment    
    states = env_info.vector_observations                       # get the current state (for the agent)
    scores = np.zeros(num_agents)                                                 # initialize the score (for the agent)
    step_num = 0
    while True:
        actions = agent.act_on_all_agents(states, add_noise)                              # select an action (for the agent)
        env_info = env.step(actions)[brain_name]                # send all actions to tne environment
        next_states = env_info.vector_observations           # get next state (for each agent)
        rewards = env_info.rewards                           # get reward (for each agent)
        dones = env_info.local_done                          # see if episode finished
        if is_training:
            is_learning_step = step_num % learn_every == 0
            agent.step(states, actions, rewards, next_states, dones, is_learning_step)
        scores += rewards                                        # update the score (for each agent)
        states = next_states                                 # roll over states to next time step
        step_num += 1
        if np.any(dones):                                               # exit loop if episode finished
            return scores

def ddpg(env, brain_name, n_episodes=2000, print_every=100, finish_score=30, learn_every=8):
    agent = Agent(state_size=33, action_size=4, random_seed=0)
    scores_deque = deque(maxlen=print_every)
    scores = []                        # All the scores
    for i_episode in range(1, n_episodes+1):
        score = np.mean(run_ddpg_for_one_episode(agent, env, brain_name, True, learn_every)) 
        scores_deque.append(score)
        scores.append(score)
        print('\rEpisode {}\tAverage Score: {:.2f}'.format(i_episode, np.mean(scores_deque)), end="")
        if i_episode % print_every == 0:
            print('\rEpisode {}\tAverage Score: {:.2f}'.format(i_episode, np.mean(scores_deque)))
        if np.mean(scores_deque)>=finish_score:
            print('\nEnvironment solved in {:d} episodes!\tAverage Score: {:.2f}'.format(i_episode-100, np.mean(scores_deque)))
            torch.save(agent.actor_local.state_dict(), 'checkpoint_actor.pth')
            torch.save(agent.critic_local.state_dict(), 'checkpoint_critic.pth')
            break
    return scores
````
